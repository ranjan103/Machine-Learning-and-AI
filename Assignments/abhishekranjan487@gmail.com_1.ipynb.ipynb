```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('./haberman.csv')
print(df.head())
```

```python
### So, its better we change the outputs to a boolean value as [1,2] could be ambiguous to us in many casess
df['status']=df['status'].map({1:True , 2:False})
print(df[df['status']==True])
```

```python
np.sum(df['status']==True)
### We have cross checked whether we the number of people of type 1 is same or not 
```

```python
### Now see the updated dataset 
print(df.head(10))
```

```python
### Percentage of people whose status is True
count_true=np.sum(df['status']==True)
count_false=np.sum(df['status']==False)
count_true=count_true/(count_true+count_false)
count_false=1-count_true
print("*************************  True People   ******************************")
print("Perentage of people whose status is True : " +  str(count_true))
print()
print("************************* False People   ******************************")
print("Percentage of people whose status is False : " + str(count_false))
```

```python
#### So, there are around 73 percent of people whose status is True and rest have status as False
```

```python
### So, As number of parameters is less than 6 , we can use pair plot 
### So, let's plot PairPlot
sns.set_style("whitegrid")
sns.pairplot(df,hue="status",size=3)
plt.show()
```

```python
### From pairplot , we can observe some patterns in it-->
### 1. Patients with more node value have less chance of survival
### 2. We can do odd one out of this plots , we can see that plots of nodes  are quite different and usefull than others
###    Rather we can say, nodes Vs year (number of operation year) can be used to separate the points among the remaining ones
```

```python
true=df.loc[df['status']==True]
false=df.loc[df['status']==False]
#print(true)

plt.plot(df['age'],np.zeros_like(df['age']),'o')
plt.plot(df['year'],np.zeros_like(df['year']),'o')
plt.plot(df['nodes'],np.zeros_like(df['nodes']),'o')
plt.show()

```

```python
### Now, let's try to plot histogram 
sns.FacetGrid(df,hue="status",size=10).map(sns.distplot,"age").add_legend()
plt.show()
```

```python
### Similar histogram of status with operation year
sns.FacetGrid(df,hue="status",size=10).map(sns.distplot,"year").add_legend()
plt.show()
```

```python
### Similar histogram with nodes
sns.FacetGrid(df,hue="status",size=10).map(sns.distplot,"nodes").add_legend()
plt.show()
```

```python
### Key points to notice --->
### 1. Those patients who survived for more than five years have nodes value around zero.
### 2. Patients whose age is more than 40 yrs (inclusive) and not much older have better chance to survive more.
```

```python
counts,bin_edges=np.histogram(true['age'],bins=10,density=True)
pdf=counts/sum(counts)
print(pdf)
cdf=np.cumsum(pdf)
### As we know , cdf is cumulative prefix sum of pdf
print(cdf)
```

```python

plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```

```python
counts1,bin_edges1=np.histogram(false,bins=10,density=True)
pdf=counts1/sum(counts1)
cdf=np.cumsum(pdf)
plt.plot(bin_edges1[1:],pdf)
plt.plot(bin_edges1[1:],cdf)
plt.show()
```
